```python
data=df[['Category','Genres','Reviews','Size','Installs','Type','Price','Content Rating','Rating']]
```

```python
data.head()
```

```python
#data.drop(["Genres"],axis=1,inplace=True)
genres_Value = data["Genres"].unique()
genres_ValueCount = len(genres_Value)
genres_dict = {}
for i in range(0,genres_ValueCount):
    genres_dict[genres_Value[i]] = i
data["Genres"] = data["Genres"].map(genres_dict).astype(int)
```

```python
from sklearn import preprocessing
#le = preprocessing.LabelEncoder()
category_list = data['Category'].unique().tolist() 
category_list = ['cat_' + word for word in category_list]
data = pd.concat([data, pd.get_dummies(data['Category'], prefix='cat')], axis=1)
category_list
```

```python
#le = preprocessing.LabelEncoder()
#df['Genres'] = le.fit_transform(df['Genres'])
```

```python
features = ['Genres','Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', ]
features.extend(category_list)
X = data[features]
y = data['Rating']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
X.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=15)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = RandomForestRegressor(n_estimators=500)
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred1=model1.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred1)
```
